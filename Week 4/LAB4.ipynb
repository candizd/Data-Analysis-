```python
import pandas as pd
data = pd.read_csv("strdata.txt",sep = "\s+")
```

```python
neu = data.Name + " " + data.Surname
data.insert(loc = 0,column = "full_name",value = neu)
```

```python
data['Counrty_code'] = data['Country'].astype('category').array.codes
```

```python
data['Counrty_code'] = data['Counrty_code'].replace([0,1,2], ['+49','+90','+44'])
data
```

```python
area =[]
tel_no = []
for i in data.Tel :
  a = str(i)[:3] 
  b = str(i)[3:10] 
  print(a)
  print(b)
  area.append(a)
  tel_no.append(b)

data['Area_Code'] = area
data['Tel_No'] = tel_no
```

```python
data
```

```python
import re
pattern = r"([A-Z0-9._%+-]+)@([A-Z0-9.-]+)\.([A-Z]{2,4})"
regex = re.compile(pattern, flags=re.IGNORECASE)
username = []
domain = []
suffix = []
for i in data['e-mail']:
  reg = regex.match(i)
  #print(reg.groups()[0])
  username.append(reg.groups()[0])
  domain.append(reg.groups()[1])
  suffix.append(reg.groups()[2])
data['Username'] = username
data['Domain'] = domain
data['Suffix'] = suffix
data
```
