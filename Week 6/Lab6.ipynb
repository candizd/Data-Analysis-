```python
import numpy as np
import pandas as pd
import re

df = pd.read_csv("/content/strdata.txt", sep = "\s+")
df
new = df["Name"] + " " +df["Surname"]
df.insert(loc = 0, column = "Fullname", value = new)
```

Fassen Sie Namen und Nachnamen in einer einzigen Spalte zusammen.

```python
df = df.drop(columns = ['Name','Surname'], axis = 1)  
df
```

```python
a = df['Country'].astype('category')
df['Country'] = df['Country'].astype('category')
df["Country"] = df['Country'].cat.codes
df
```

Konvertieren Sie "Land" in einen kategorischen Datentyp und kodieren Sie dann die Länder als Ganzzahlen (z. B. Telefonvorwahlen).

```python
a.replace(['Germany', 'Turkey', 'UK'], ['+49', '+90', '+44'], inplace = True)
a 
df['Country_Code'] = a
df['Tel'] = df['Country_Code'] + df['Tel'].astype(str) 
df
```

 Zerlegen Sie die Telefonnummern in Ortsvorwahl und Rufnummer.

```python
df['Area Code'] = df['Tel'].str.slice(start = 3, stop = 6, step = 1)
df
df['Tel No.'] = df['Tel'].str.slice(start = 7, stop = 13, step = 1)
df
```

Aufteilen (mit Regexp) von Benutzernamen, Domänennamen und Suffixen in separate Spalten.

```python
pattern = r"([A-Z0-9._%+-]+)@([A-Z0-9.-]+)\.([A-Z]{2,4})"
regex = re.compile(pattern, flags=re.IGNORECASE)
username = df['e-mail'].str.findall(pattern, flags=re.IGNORECASE).str[0]
df['Username'] = username.str.get(0)
df['Domain'] = username.str.get(1)
df['Suffix'] = username.str.get(2)
df
```
